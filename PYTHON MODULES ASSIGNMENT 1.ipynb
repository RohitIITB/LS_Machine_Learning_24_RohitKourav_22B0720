```python
#Task: Perform the dot product of I and 9I+1 using numpy, here I is referred to as an 3x3 Identity matrix.

I = np.identity(3, dtype=int)
dot_Product = np.matmul(I,(9*I+1))
print(dot_Product)
```

```python
import pandas as pd

## Create a DataFrame from a dictionary

data = {
    'Name': ['Ramesh', 'Mahesh', 'Suresh'],
    'Age': [25, 30, 35],
    'City': ['Bangalore', 'Mumbai', 'Delhi']
}
df = pd.DataFrame(data)

#Display the first 2 rows of the data frame

print(df.head(2))

#Print the age column

print(df.iloc[:,1])

#Filter rows where age is greater than 26

print("age greater than 26")
print(df[df['Age']>26])

#Add a new column 'Country' with the value 'India' for all rows

df['Country']=['India', 'India', 'India']
print(df)
data1 = {
    'Name': ['Ramesh', 'Mahesh', 'Suresh'],
    'Age': [25, None, 35],
    'City': ['Bangalore', 'Mumbai', 'Delhi']
}
df2= pd.DataFrame(data1)

# Fill missing values in the 'Age' column with the mean age

mean_age = df2['Age'].mean()
print("mean age is " + str(mean_age))
df2['Age'].fillna(mean_age, inplace=True)
print(df2)
```

```python
import matplotlib.pyplot as plt
import numpy as np #numpy module

xpoints=np.array([1,2,3,4])
ypoints=np.array([2,4,6,8])

#Plot these points without drawing a line
plt.figure(figsize=(4, 3))
plt.scatter(xpoints, ypoints, color='blue')
plt.grid(True)
plt.show()

#Plotting with marker: Plot these points with a marker(Star marker)
plt.figure(figsize=(4, 3))
plt.scatter(xpoints, ypoints, color='red', marker='*') 
plt.grid(True)
plt.show()

#Using fmt format, add circular marker,red color and Dashed line
plt.figure(figsize=(4, 3))
plt.plot(xpoints, ypoints, 'ro--')
plt.show()

#Add xlabel,ylabel and title for the plot.
plt.figure(figsize=(4, 3))
plt.plot(xpoints, ypoints, 'ro--')
plt.xlabel('xpoints')
plt.ylabel('ypoints')
plt.title('X VS Y')
plt.show()

#Create a scatter plot for xpoints and ypoints
plt.figure(figsize=(4, 3))
plt.scatter(xpoints, ypoints, c=['blue', 'green', 'red', 'yellow'])
plt.xlabel('xpoints')
plt.ylabel('ypoints')
plt.title('Scatter Plot with Specific Colors')
plt.grid(True)
plt.show()

#Set color to the scatter plot. Blue,Green,Red and yellow color for each point respectively
```

```python
import random
import numpy as np

# Set the seed of random to 20 (using NumPy)
np.random.seed(20)

arr1 = np.array([1, 24, 31, 45, 73, 81, 94, 25])

#Using the random module pick 4 different random numbers from arr1 and return their sum.
random_indices = np.random.choice(len(arr1), size=4, replace=False)
# I have used len so numbers should be in range of array indices otherwise creationg problems
random_numbers = arr1[random_indices]

# Calculate their sum
sum_random_numbers = np.sum(random_numbers)
print("Random indices are ", random_indices)
print("Random numbers picked:", random_numbers)
print("Sum of random numbers:", sum_random_numbers)
```
